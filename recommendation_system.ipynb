```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
column_names=['user_id','item_id','rating','time_stamp']
df=pd.read_csv('u.data.txt', sep='\t',names=column_names)
```

```python
df.head(5)
```

```python
movie_titles=pd.read_csv('Movie_Id_Titles.txt')
movie_titles.head()
```

```python
df=pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
import seaborn as sns
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating=pd.DataFrame(df.groupby('title')['rating'].mean())
rating.head()
```

```python
rating['num of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
rating.head()
```

```python
plt.figure(figsize=(10,5))
plt.hist(rating['num of ratings'])
```

```python
plt.figure(figsize=(10,5))
plt.hist(rating['rating'])
```

```python
moviemat=df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
rating.sort_values('num of ratings',ascending=False).head()
```

```python
starwars_user_rating=moviemat['Star Wars (1977)']
starwars_user_rating.head()
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_rating)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['correlations'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('correlations',ascending=False).head()
```

```python
corr_starwars=corr_starwars.join(rating['num of ratings'])
corr_starwars
```

```python
corr_starwars[corr_starwars['num of ratings']>=100].sort_values('correlations',ascending=False).head()
```
